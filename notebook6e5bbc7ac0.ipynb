```python
import time
from datetime import datetime
import csv
import sys

name = input("Please enter your name: ")

filename = f"{name}_entries.csv"
with open(filename, "w", newline='') as f1:  
    writer = csv.writer(f1)
    writer.writerow(["Name", "Chosen Number", "Time delay (seconds)", "Timestamp"]) 

    while True:
        start = time.time()
        
        while True:
            num_input = input("Please enter your chosen number(1-9): ")
            
            if num_input.lower() == "quit":
                print("You messed up")
                num = "quit" 
                break 
            
            try:
                num = int(num_input)
                if 1 <= num <= 9:
                    break
                else:
                    print("Come on man, please enter a valid integer between 1 and 9.")
            except ValueError:
                print("Invalid input. Please enter a valid integer")

        if num == "quit":
            break

        final = time.time()
        writer.writerow([name, num, final - start, datetime.now().strftime("%Y-%m-%d %H:%M:%S")])
```
